```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


dataset = pd.read_csv('../dataset/fake_house_price_data.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1:].values

# scatter graph for the data
plt.scatter(x,y,marker="x",s = 20,c="red")
plt.xlim(x.min() * 0.8, x.max() * 1.1)
plt.ylim(y.min() * 0.8, y.max() * 1.1)
plt.ylabel("Price($) in 1000's")
plt.xlabel("Size in feet^2")
# draw approximate line to fit the data
lx = np.arange(800,2200)
ly = 0.20 * lx - 0;
plt.plot(lx,ly,c = 'blue')
plt.show()
```

```python
# draw cost function image for the data
theta_1 = np.arange(0.1,0.3,0.005)
cost = 1/2 * (theta_1*x - y ) ** 2
ycost = np.sum(cost,axis=0)
figure = plt.figure()
plt.plot(theta_1,ycost,c = 'red')
plt.xlabel('Theta_1');
plt.ylabel('Cost')
plt.suptitle('Cost function', fontsize=16)
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
fig = plt.figure()
plt.suptitle('Cost')
theta_1= np.arange(0.1,0.3,0.01)
theta_0= np.arange(0,50)
X,Y = np.meshgrid(theta_1,theta_0)
def flat(t1,t0):
    return (1/2*(t1*x - t0 - y)**2).sum()
z = np.array([flat(a,b) for a,b in zip(np.ravel(X),np.ravel(Y))])
Z = z.reshape(X.shape)
ax = plt.gca(projection='3d')
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=cm.coolwarm)
ax.set_zlabel('Cost Values')
ax.set_xlabel('Theta_1')
ax.set_ylabel('Theta_0')
plt.tight_layout()
```
